```python
def classify_row(row):

    protein = str(row["Protein"]) if pd.notna(row["Protein"]) else ""
    mapped = str(row["Mapped Proteins"]) if pd.notna(row["Mapped Proteins"]) else ""

    # join all hits
    all_hits = protein
    if mapped.strip():
        all_hits += "," + mapped

    has_spliced = "LIEPE-SPLICED" in all_hits
    has_not_spliced = "LIEPE-NOT-SPLICED" in all_hits

    has_swissprot = "sp|" in all_hits and not (
        all_hits.startswith("sp|LIEPE")
    )

    has_trembl = "tr|" in all_hits

    mapped_empty = mapped.strip() == ""

    # classification

    if has_spliced and not has_swissprot and not has_trembl and not has_not_spliced:
        return "Spliced_only"

    elif has_spliced and (has_swissprot or has_trembl or has_not_spliced):
        return "Spliced_with_alternative"

    elif has_not_spliced and not has_swissprot and not has_trembl:
        return "Liepe_not_spliced_only"

    elif has_swissprot and not has_spliced:
        return "Canonical"

    elif has_trembl and not has_swissprot and not has_spliced:
        return "TrEMBL_only"

    else:
        return "Other"


psm["Mapping_status"] = psm.apply(classify_row, axis=1)
psm.head(2)
```

```python
def contains_spliced(text):
    return "LIEPE-SPLICED" in str(text)

def contains_not_spliced(text):
    return "LIEPE-NOT-SPLICED" in str(text)

def contains_trembl(text):
    return "tr|" in str(text)

def contains_real_swissprot(text):

    text = str(text)

    hits = [x.strip() for x in text.split(",")]

    for h in hits:

        if h.startswith("sp|LIEPE"):
            continue

        if h.startswith("sp|"):
            return True

    return False


combined = (
    psm["Protein"].fillna("")
    + ","
    + psm["Mapped Proteins"].fillna("")
)

psm["has_spliced"] = combined.apply(contains_spliced)
psm["has_not_spliced"] = combined.apply(contains_not_spliced)
psm["has_trembl"] = combined.apply(contains_trembl)
psm["has_swissprot"] = combined.apply(contains_real_swissprot)
psm["mapped_empty"] = (
    psm["Mapped Proteins"]
    .fillna("")
    .str.strip()
    .eq("")
)

psm.head(3)
```

```python
def classify_row_with_first_hit_priority(row):

    protein = str(row["Protein"]) if pd.notna(row["Protein"]) else ""
    mapped = str(row["Mapped Proteins"]) if pd.notna(row["Mapped Proteins"]) else ""

    # join all hits
    is_spliced = "LIEPE-SPLICED" in protein
    is_not_spliced = "LIEPE-NOT-SPLICED" in protein
    is_swissprot = "sp|" in protein and not (
        protein.startswith("sp|LIEPE")
    )
    is_trembl = "tr|" in protein

    mapped_empty = pd.isna(mapped) or str(mapped).strip() == ""
    mapped_spliced = "LIEPE-SPLICED" in mapped
    mapped_not_spliced = "LIEPE-NOT-SPLICED" in mapped
    mapped_swissprot = "sp|" in mapped and not (
        mapped.startswith("sp|LIEPE")
    )
    mapped_trembl = "tr|" in mapped


    # classification
    if is_not_spliced:
        return "Liepe_not_spliced"
    elif is_swissprot:
        return "Swiss-Prot"
    elif is_trembl:
        return "TrEMBL"
    elif is_spliced:
        if mapped_empty:
            return "Spliced_only"
        elif mapped_spliced and not mapped_swissprot and not mapped_trembl:
            return "Spliced_with_LIEPE-SPLICED_alternative"
        elif mapped_not_spliced and not mapped_swissprot and not mapped_trembl:
            return "Spliced_with_LIEPE-NOT-SPLICED_alternative"
        elif mapped_swissprot and not mapped_spliced and not mapped_not_spliced and not mapped_trembl:
            return "Spliced_with_Swiss-Prot_alternative"
        elif mapped_trembl and not mapped_spliced and not mapped_not_spliced and not mapped_swissprot:
            return "Spliced_with_TrEMBL_alternative"
        elif mapped_swissprot and mapped_trembl and not mapped_spliced and not mapped_not_spliced:
            return "Spliced_with_Uniprot_alternative"
        else:
            return "Spliced_with_alternative"
    else:
        return "Other"
    
```

```python
psm_filtered["Class"] = psm_filtered.apply(classify_row_with_first_hit_priority, axis=1)
psm_filtered.head(15)
```
